```python
# Dependencies & Setup
import pandas as pd
import numpy as mp
import datetime as dt
```

```python
# Files to load
jan2020 = "Monthly_Data_2020/JC-202001-citibike-tripdata.csv"
feb2020 = "Monthly_Data_2020/JC-202002-citibike-tripdata.csv"
mar2020 = "Monthly_Data_2020/JC-202003-citibike-tripdata.csv"
apr2020 = "Monthly_Data_2020/JC-202004-citibike-tripdata.csv"
may2020 = "Monthly_Data_2020/JC-202005-citibike-tripdata.csv"
jun2020 = "Monthly_Data_2020/JC-202006-citibike-tripdata.csv"
jul2020 = "Monthly_Data_2020/JC-202007-citibike-tripdata.csv"
aug2020 = "Monthly_Data_2020/JC-202008-citibike-tripdata.csv"
sep2020 = "Monthly_Data_2020/JC-202009-citibike-tripdata.csv"
oct2020 = "Monthly_Data_2020/JC-202010-citibike-tripdata.csv"
nov2020 = "Monthly_Data_2020/JC-202011-citibike-tripdata.csv"
dec2020 = "Monthly_Data_2020/JC-202012-citibike-tripdata.csv"

# Read 2020 Citi Bike Data Files & Store into Pandas DataFrame
jan2020_df = pd.read_csv(jan2020, encoding="ISO-8859-1")
feb2020_df = pd.read_csv(feb2020, encoding="ISO-8859-1")
mar2020_df = pd.read_csv(mar2020, encoding="ISO-8859-1")
apr2020_df = pd.read_csv(apr2020, encoding="ISO-8859-1")
may2020_df = pd.read_csv(may2020, encoding="ISO-8859-1")
jun2020_df = pd.read_csv(jun2020, encoding="ISO-8859-1")
jul2020_df = pd.read_csv(jul2020, encoding="ISO-8859-1")
aug2020_df = pd.read_csv(aug2020, encoding="ISO-8859-1")
sep2020_df = pd.read_csv(sep2020, encoding="ISO-8859-1")
oct2020_df = pd.read_csv(oct2020, encoding="ISO-8859-1")
nov2020_df = pd.read_csv(nov2020, encoding="ISO-8859-1")
dec2020_df = pd.read_csv(dec2020, encoding="ISO-8859-1")
```

```python
# Combine All 12 2020 dataframes Into Single DataFrame
combined_2020_df = jan2020_df.append([feb2020_df, mar2020_df, apr2020_df, may2020_df, jun2020_df, jul2020_df, aug2020_df, sep2020_df, oct2020_df, nov2020_df, dec2020_df], ignore_index=True)
combined_2020_df.head()
```

```python
# Rename Columns
combined_2020_df = combined_2020_df.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
combined_2020_df.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
combined_2020_df["Start Time"] = pd.to_datetime(combined_2020_df["Start Time"])
combined_2020_df["Stop Time"] = pd.to_datetime(combined_2020_df["Stop Time"])
combined_2020_df.Gender.replace([0, 1, 2], ['Unknown','Male', 'Female'], inplace=True)
combined_2020_df.head()
```

```python
# Save Combined DataFrames to CSV
combined_2020_df.to_csv("Citi_Bike_Data_2020/year-2020-citibike-tripdata.csv", index=False, header=True)
```

```python
# Sort and display the top five performing schools by % overall passing.
sorted_combined_2020_df = combined_2020_df.sort_values("Start Time", ascending=True)
sorted_combined_2020_df.head()
```

```python
sorted_combined_2020_df.to_csv("Citi_Bike_Data_2020/sorted-year-2020-citibike-tripdata.csv", index=False, header=True)
```
